```python
import numpy as np
import pylab as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
def running_mean(x, N):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
```

```python
stats = np.load('./stats/52-rewards.npy')
losses = np.load('./stats/52-losses.npy')
#print(stats)
```

```python
stats_np = np.asarray(stats)
fig=plt.figure(figsize=(6,3.8))
plt.plot(stats_np[:10000,1], label="return")
plt.plot(running_mean(stats_np[:10000,1],500), label="smoothed-return")
plt.legend()
```
